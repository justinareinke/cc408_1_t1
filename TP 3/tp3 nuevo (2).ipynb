Implementen los siguientes métodos reportando luego la matriz de
confusión, la curva ROC, los valores de AUC y de Accuracy de cada uno:
● Regresión logística
● Análisis discriminante lineal
● KNN con k=3
● Naive Bayes

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, auc, RocCurveDisplay
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Seleccionamos solo las variables especificadas
variables = ['ch04', 'ch06', 'ch07', 'ch08', 'nivel_ed', 'estado', 'cat_inac', 'ipcf']

# Filtramos las variables en el conjunto de datos respondieron para 2004 y 2024
respondieron_2004 = respondieron[respondieron['ano4'] == 2004][variables + ['desocupado']]
respondieron_2024 = respondieron[respondieron['ano4'] == 2024][variables + ['desocupado']]

# Dividimos en conjuntos de entrenamiento y prueba (70% entrenamiento, 30% prueba) con semilla 101 para 2004
train_2004, test_2004 = train_test_split(respondieron_2004, test_size=0.3, random_state=101)

# Dividimos en conjuntos de entrenamiento y prueba (70% entrenamiento, 30% prueba) con semilla 101 para 2024
train_2024, test_2024 = train_test_split(respondieron_2024, test_size=0.3, random_state=101)

# Establecemos la variable dependiente (y) y variables independientes (X) para cada conjunto de entrenamiento y prueba
y_train_2004 = train_2004['desocupado']
X_train_2004 = train_2004[variables]

y_test_2004 = test_2004['desocupado']
X_test_2004 = test_2004[variables]

y_train_2024 = train_2024['desocupado']
X_train_2024 = train_2024[variables]

y_test_2024 = test_2024['desocupado']
X_test_2024 = test_2024[variables]

# Convertir variables categóricas a dummy variables
X_train_2004 = pd.get_dummies(X_train_2004, drop_first=True)
X_test_2004 = pd.get_dummies(X_test_2004, drop_first=True)

X_train_2024 = pd.get_dummies(X_train_2024, drop_first=True)
X_test_2024 = pd.get_dummies(X_test_2024, drop_first=True)

# Misma cantidad de columnas 
X_test_2004 = X_test_2004.reindex(columns=X_train_2004.columns, fill_value=0)
X_test_2024 = X_test_2024.reindex(columns=X_train_2024.columns, fill_value=0)

# Agregar columna de unos  a X_train y X_test para cada año
X_train_2004.insert(0, 'Intercept', 1)
X_test_2004.insert(0, 'Intercept', 1)

X_train_2024.insert(0, 'Intercept', 1)
X_test_2024.insert(0, 'Intercept', 1)

# Lista para almacenar resultados
results = []

# Definir función para entrenar y evaluar modelos
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    # Entrenar el modelo
    model.fit(X_train, y_train)
    # Predecir en el conjunto de prueba
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
    
    # Calcular métricas
    conf_matrix = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba) if y_pred_proba is not None else (None, None, None)
    auc_score = auc(fpr, tpr) if y_pred_proba is not None else None
    
    # Guardar resultados
    results.append({
        "Model": model_name,
        "Confusion Matrix": conf_matrix,
        "Accuracy": accuracy,
        "AUC": auc_score
    })
    
    # Imprimimos resultados
    print(f"\n{model_name}")
    print("Confusion Matrix:\n", conf_matrix)
    print(f"Accuracy: {accuracy}")
    if auc_score is not None:
        print(f"AUC: {auc_score}")
    
    # Curva ROC
    if y_pred_proba is not None:
        RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
        plt.title(f"ROC Curve - {model_name}")
        plt.show()

# 1. Regresión logística
log_reg = LogisticRegression()
evaluate_model(log_reg, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "Logistic Regression")

# 2. Análisis discriminante lineal
lda = LinearDiscriminantAnalysis()
evaluate_model(lda, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "Linear Discriminant Analysis")

# 3. K-Nearest Neighbors (k=3)
knn = KNeighborsClassifier(n_neighbors=3)
evaluate_model(knn, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "K-Nearest Neighbors (k=3)")

# 4. Naive Bayes
nb = GaussianNB()
evaluate_model(nb, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "Naive Bayes")

# Mostrar resumen de resultados
for result in results:
    print(f"\nModel: {result['Model']}")
    print("Confusion Matrix:\n", result["Confusion Matrix"])
    print("Accuracy:", result["Accuracy"])
    if result["AUC"] is not None:
        print("AUC:", result["AUC"])
```

3.  Compare los resultados de 2004 versus 2024. ¿Cuál de los métodos predice mejor en cada año? Justifiquen detalladamente utilizando las medidas de precisión que conocen. 

```python
from sklearn.metrics import mean_squared_error
#repito el procedimiento de arriba pero le agrego medida de performance de ECM para datos de testeo 

# Defininimos función para entrenar, evaluar y calcular ECM para los modelos
def evaluate_model_with_mse(model, X_train, y_train, X_test, y_test, model_name, year):
    # Entrenar el modelo
    model.fit(X_train, y_train)
    # Predecir en el conjunto de prueba
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
    
    # Calculamos los valores para evlauar la perfomrance de los distintos modelos 
    conf_matrix = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba) if y_pred_proba is not None else (None, None, None)
    auc_score = auc(fpr, tpr) if y_pred_proba is not None else None
    
    # Guardar resultados
    results.append({
        "Model": model_name,
        "Year": year,
        "Confusion Matrix": conf_matrix,
        "Accuracy": accuracy,
        "AUC": auc_score,
        "MSE": mse
    })
    
    # Imprimir resultados
    print(f"\n{model_name} - {year}")
    print("Confusion Matrix:\n", conf_matrix)
    print(f"Accuracy: {accuracy}")
    print(f"MSE: {mse}")
    if auc_score is not None:
        print(f"AUC: {auc_score}")
    
    # Curva ROC
    if y_pred_proba is not None:
        RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
        plt.title(f"ROC Curve - {model_name} - {year}")
        plt.show()

# Lista para almacenar resultados
results = []

# Evaluar modelos para 2004 y 2024
models = [
    ("Logistic Regression", LogisticRegression()),
    ("Linear Discriminant Analysis", LinearDiscriminantAnalysis()),
    ("K-Nearest Neighbors (k=3)", KNeighborsClassifier(n_neighbors=3)),
    ("Naive Bayes", GaussianNB())
]

# Iterar sobre los modelos y evaluar en 2004 y 2024
for model_name, model in models:
    # Evaluar en 2004
    evaluate_model_with_mse(model, X_train_2004, y_train_2004, X_test_2004, y_test_2004, model_name, "2004")
    # Evaluar en 2024
    evaluate_model_with_mse(model, X_train_2024, y_train_2024, X_test_2024, y_test_2024, model_name, "2024")

# Comparación de resultados
print("\nResumen de Resultados Comparativos (2004 vs 2024):")
for result in results:
    print(f"\nModel: {result['Model']} - Year: {result['Year']}")
    print("Confusion Matrix:\n", result["Confusion Matrix"])
    print("Accuracy:", result["Accuracy"])
    print("AUC:", result["AUC"])
    print("MSE:", result["MSE"])
```
